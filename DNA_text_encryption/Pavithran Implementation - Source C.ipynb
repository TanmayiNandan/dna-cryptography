```python
def convert_to_binary_8dig(n):
    n = str(n)
    zeroes = 0
    for i in range(0,len(n)):
        if n[i]=='0':
            zeroes+=1
            continue
        break
    n = int(n)
    # print(f'Zeroes: {zeroes}')
    binary = bin(n).replace("0b", "")
    diff = 8 - len(binary)
    if diff>0:
        for i in range(0,diff):
            binary = '0'+binary
    # print(binary)
    return binary

def convert_str_to_ascii(text):
    ascii_array = []
    for x in list(text):
        ascii_array.append(ord(x))
    return ascii_array

def xor(A,B):
    if A=='0' and B=='0':
        return '0'
    if A=='0' and B=='1':
        return '1'
    if A=='1' and B=='0':
        return '1'
    if A=='1' and B=='1':
        return '0'

def construct_dna_binary_table():
    from itertools import permutations
    x = 'ATCG'
    perms = []
    for i in range(1, len(x)+1):
        for c in permutations(x, i):
            val = "".join(c)
            if len(val)==4:
                perms.append(val)
    dna_binary = {}
    for i in range(0,len(perms)):
        dna_binary[i]=perms[i]
    # print(dna_binary)
    return (dna_binary)


state_table = [
    {'A': 1, 'T': 2, 'C': 0, 'G': 3},
    {'A': 3, 'T': 1, 'C': 2, 'G': 0},
    {'A': 0, 'T': 3, 'C': 1, 'G': 2},
    {'A': 2, 'T': 0, 'C': 3, 'G': 1}
]
output_table = [
    {'A': 'T', 'T': 'A', 'C': 'G', 'G': 'C'},
    {'A': 'C', 'T': 'T', 'C': 'A', 'G': 'G'},
    {'A': 'G', 'T': 'C', 'C': 'T', 'G': 'A'},
    {'A': 'A', 'T': 'G', 'C': 'C', 'G': 'T'},
]
```

```python
def source_C_encryption(plaintext, RASK):
    # a) The sender selects the plaintext ‘hello’.
    
    # b) Convert each character of plaintext to its ASCII value.
    plaintext_ascii = convert_str_to_ascii(plaintext)
    
    # c) Convert ASCII values to its equivalent binary value to form a binary string
    plaintext_binary = ''
    for val in plaintext_ascii:
        plaintext_binary += convert_to_binary_8dig(val)
    # print(plaintext_binary)
    
    # d) Since the length is less than 256, RASK is truncated by taking the last 40 bits
    if len(plaintext_binary)<len(RASK):
        RASK = RASK[len(RASK)-len(plaintext_binary):]
    
    if len(plaintext_binary)>len(RASK):
        mod_RASK = ''
        q = int(len(plaintext_binary)/len(RASK))
        r = len(plaintext_binary)%len(RASK)
        for i in range(0,q):
            mod_RASK = mod_RASK+RASK
        for i in range(0,r):
            mod_RASK = mod_RASK+RASK[i]
        RASK = mod_RASK
    
    # e) EXOR operation is performed between the binary string and the truncated RASK.
    plaintext_xor = ''
    for i in range(0,len(plaintext_binary)):
        plaintext_xor += xor(plaintext_binary[i],RASK[i])
    # print(plaintext_xor)
    
    # f) Convert each two bits of the binary number obtained from (e) to the corresponding DNA base as per row number 0 of Table 4.
    row_0 = construct_dna_binary_table()[0]
    dna_conversion  = {'00':row_0[0], '01':row_0[1], '10':row_0[2], '11':row_0[3]}
    dnaD1 = ''
    for i in range(0,len(plaintext_xor),2):
        dnaD1 += dna_conversion[plaintext_xor[i:i+2]]
    # print(dnaD1)
    
    # g) dnaD1 is given as input to the Mealy machine to generate a new DNA sequence i.e., dnaD2.
    # Mealy Machine
    state = 0
    dnaD2 = ''
    for letter in list(dnaD1):
        dnaD2 += output_table[state][letter]
        state = state_table[state][letter]
    # print(dnaD2)
    
    # h) Reverse dnaD2 to generate another DNA sequence, which is the final ciphertext.
    ciphertext = dnaD2[::-1]

    return (ciphertext)
```

```python
def source_C_decryption(ciphertext, RASK):
    # a) The receiver reverses the ciphertext to get a DNA sequence i.e. dnaD2.
    dnaD2 = ciphertext[::-1]
    
    # b) Then, dnaD2 is given as input to the Mealy machine to gen- erate another DNA sequence i.e. dnaD1
    state = 0
    dnaD1 = ''
    for letter in list(dnaD2):
        dnaD1 += output_table[state][letter]
        state = state_table[state][output_table[state][letter]]
    # print(dnaD1)
    
    # c) dnaD1 is converted into its equivalent binary number as per the row number 0 of Table 4
    dna_binary = ''
    row_0 = construct_dna_binary_table()[0]
    dna_conversion  = {'00':row_0[0], '01':row_0[1], '10':row_0[2], '11':row_0[3]}
    for letter in list(dnaD1):
        dna_binary += list(dna_conversion.keys())[list(dna_conversion.values()).index(letter)]
    # print(dna_binary)
    
    # d) Since the length is less than 256-bit, RASK is truncated by taking the last 40 bits
    if len(dna_binary)<len(RASK):
        RASK = RASK[len(RASK)-len(dna_binary):]
    if len(dna_binary)>len(RASK):
        mod_RASK = ''
        q = int(len(dna_binary)/len(RASK))
        r = len(dna_binary)%len(RASK)
        for i in range(0,q):
            mod_RASK = mod_RASK+RASK
        for i in range(0,r):
            mod_RASK = mod_RASK+RASK[i]
        RASK = mod_RASK
    
    # e) EXOR operation is performed between the resultants of the above two steps.
    plaintext_xor = ''
    # print(RASK)
    for i in range(0,len(dna_binary)):
        plaintext_xor += xor(dna_binary[i],RASK[i])
    # print(plaintext_xor)
    
    # f) The binary number is converted into its equivalent ASCII values.
    ascii_values = []
    while len(plaintext_xor)>8:
        ascii_values.append(int(plaintext_xor[:8],2))
        plaintext_xor = plaintext_xor[8:]
    ascii_values.append(int(plaintext_xor,2))
    # print(ascii_values)
    
    # g) At last, the ASCII values are converted to their equivalent characters to retrieve the plaintext
    plaintext = ''
    for i in ascii_values:
        plaintext += chr(i)
    return plaintext
```
