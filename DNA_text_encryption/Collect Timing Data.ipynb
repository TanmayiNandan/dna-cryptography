```python
import time
import pandas as pd
import csv

def gather_ed_times(plaintext_length, num_cycles_each):

    num_words = []
    encryption_time = []
    decryption_time = []
    alg_run = []


    for this_input_length in plaintext_length:
        print('STARTING INPUTS OF LENGTH ', this_input_length)
    
        for iteration in range(0, num_cycles_each):

            num_words.append(this_input_length)
    
            selected_plaintext = get_plaintext(n_input_words = this_input_length) 
    
            # print('STARTING ITERATION :', iteration)
            # print('plaintext :', selected_plaintext)
            # print()
            
            # print('SOURCE B')
    
            # SOURCE B BEGIN
            encryption_start_time = time.time()
            generated_ciphertext = encryption_process(m = selected_plaintext, 
                                                      num_round_functions = 10)
            encryption_time.append(time.time() - encryption_start_time)
            
            # print('plaintext :', selected_plaintext)        
            # print('ciphertext', generated_ciphertext)
    
            decryption_start_time = time.time()
            computed_plaintext = decryption_process(ciphertext = generated_ciphertext, 
                                                    num_round_functions = 10)
            decryption_time.append(time.time() - decryption_start_time)
            
            alg_run.append('source_b')
    
            # print('computed plaintext :', computed_plaintext)
            # print()
    
            
            num_words.append(this_input_length)

            # SOURCE C BEGIN 
            # print('SOURCE C')
            key = '0110000101100001011001010110001001111001011000110111000001100100011100000110010100111000011001100011010001100111001100110110100000110110010000000111001001100111011011010110000101101001011011000010111001100011011011110110110100110000011000010010110100110001'
            encryption_start_time = time.time()
            generated_ciphertext = source_C_encryption(plaintext = selected_plaintext, RASK = key)
            encryption_time.append(time.time() - encryption_start_time)
    
            # print('ciphertext', generated_ciphertext)
    
            decryption_start_time = time.time()
            computed_plaintext = source_C_decryption(ciphertext = generated_ciphertext, RASK = key)        
            decryption_time.append(time.time() - decryption_start_time)
            
            alg_run.append('source_c')
    
            # print('computed plaintext :', computed_plaintext)
            
            num_words.append(this_input_length)

            # SOURCE E BEGIN
            # print('SOURCE E')
            encryption_start_time = time.time()
            generated_ciphertext, modified_key = source_E_encryption(plaintext = selected_plaintext, key = 'ACTGACTGA')
            encryption_time.append(time.time() - encryption_start_time)   
            
            # print('ciphertext', generated_ciphertext)
            gen_modified_key = modified_key
    
            decryption_start_time = time.time()
            computed_plaintext = source_E_decryption(ciphertext = generated_ciphertext, 
                                                     modified_key = gen_modified_key)
            decryption_time.append(time.time() - decryption_start_time)
            
            alg_run.append('source_e')
    
            # print('computed plaintext :', computed_plaintext)

            print('COMPLETED ITERATION ', iteration)
                
    df = pd.DataFrame(data={"num_words": num_words,
                            "encryption_time": encryption_time, 
                            "decryption_time": decryption_time,
                            "alg_run": alg_run})
    print(df)
    filename = f'{this_input_length}.csv' 
    # filename = '1600_2000.csv'
    df.to_csv(filename, sep=',',index=False)
    
    print('COMPLETED')        
    return [encryption_time, decryption_time]
```
